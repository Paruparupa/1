#### NumPy用于处理数据的各种方法


5位同学参加了学业水平考试，考试科目共7门





#### 创建names 姓名ndarray一维数组

```python
import numpy as np

# 创建姓名的一维数组
names = np.array(['王微', '肖良英', '方绮雯', '刘旭阳', '钱易铭'])

# 打印数组以验证
print("Names:", names)
```

#### 创建subjects 课程ndarray一维数组

```python
# 创建课程的一维数组
subjects = np.array(['Math', 'English', 'Python', 'Chinese', 'Art', 'Database', 'Physics'])

# 打印数组以验证
print("Subjects:", subjects)
```

#### 创建socres 分数ndarray二维数组

```python
# 创建分数的二维数组
scores = np.array([
    [70, 85, 77, 90, 82, 84, 89],
    [60, 64, 80, 75, 80, 92, 90],
    [90, 93, 88, 87, 86, 90, 91],
    [80, 82, 91, 88, 83, 86, 80],
    [88, 72, 78, 90, 91, 73, 80]
])

# 打印数组以验证\
print("Scores:\n", scores)
```

#### 切片相关操作（可扩充单元格）

```python
# 提取王微的分数
wangwei_scores = scores[0, :]
print("王微的分数:", wangwei_scores)

# 提取所有学生的Math和Python分数
math_python_scores = scores[:, [0, 2]]
print("所有学生的Math和Python分数:\n", math_python_scores)

# 提取第二行（肖良英）的分数
xiao_scores = scores[1, :]
print("肖良英的分数:", xiao_scores)

# 提取第三列（Python）的所有学生的分数
python_scores = scores[:, 2]
print("所有学生的Python分数:", python_scores)

# 提取从第二行第二列开始的2x2区域的分数
sub_matrix = scores[1:3, 1:3]
print("2x2子矩阵:\n", sub_matrix)

# 提取所有学生的分数，步长为2（即每隔一个元素取一个）
scores_stride = scores[:, ::2]
print("步长为2的分数:\n", scores_stride)
```

#### 1、使用布尔条件选择肖良英和方绮雯两个学生的成绩

```python
selected_students = ((names == '肖良英') | (names == '方绮雯'))
print(selected_students)
print(scores[selected_students, :])
```

#### 2、使用布尔条件选择“肖良英”和“方绮雯”的“Math”和“Python”课程成绩

```python
selected_students = ((names == '肖良英') | (names == '方绮雯'))
selected_subjects = ((subjects == 'Math') | (subjects == 'Python'))
print(selected_students)
print(selected_subjects)
print(scores[selected_students, :][:, selected_subjects])
```

#### 3、为所有同学的所有课程成绩增加5分

```python
scores += 5
print(scores)
```

#### 4、每个科目基础分不同，为各科目增加相应的基础分，3，4，5，3，6，7，2

```python
# 创建一个包含基础分的数组
base_scores = np.array([3, 4, 5, 3, 6, 7, 2])

# 将基础分加到scores数组的每一列上
adjusted_scores = scores + base_scores

print("增加基础分后的成绩:\n", adjusted_scores)
```

#### 5、修改肖良英英语成为90

```python
selected_students = (names == '肖良英')
selected_subjects = (subjects == 'English')
print(selected_students)
print(selected_subjects)
scores[selected_students, selected_subjects] = 90
print(scores)
```

#### 6、使用floor函数将同学的考试成绩转换成整数形式的十分制分数。


```python
scores_tenth = np.floor(scores / 10)

print("转换成十分制分数的成绩:\n", scores_tenth)
```

#### 7、使用subtract函数将每个同学的分数减去3分。

```python
scores_subtract_3 = np.subtract(scores, 3)

print("减去3分后的成绩:\n", scores_subtract_3)
```

#### 8、统计不同科目的成绩总分


```python
print(scores.sum(axis=0))
```

#### 9、求“王微”所有课程成绩的平均分


```python
# 找到王微在names数组中的索引
wangwei_index = np.where(names == '王微')[0][0]
print(wangwei_index)

# 提取王微的成绩
wangwei_scores = scores[wangwei_index, :]
print(wangwei_scores)

# 计算王微所有课程成绩的平均分
wangwei_average = np.mean(wangwei_scores)

print("王微所有课程成绩的平均分:", wangwei_average)
```

#### 10、查询英语考试成绩最高同学的姓名


```python
# 提取所有学生的英语成绩
english_scores = scores[:, 1]

# 找到英语成绩最高的索引
highest_english_index = np.argmax(english_scores)

# 使用索引获取最高英语成绩同学的姓名
highest_english_student = names[highest_english_index]

print("英语考试成绩最高的同学的姓名:", highest_english_student)
```

#### 11、使用布尔条件选择并显示subjects数组中的英语和物理科目名称。


```python
# 使用布尔数组选择科目名称
selected_subjects = subjects[(subjects=='English')|(subjects=='Physics')]

print("选择的科目名称:", selected_subjects)
```

#### 12、选择并显示scores数组中行号2、4同学的数学和Python成绩。

```python
print(scores[[2,4], (subjects=='Python')])
```

#### 13、选择并显示scores数组中“王微”和“刘旭阳”的英语和艺术课程成绩。


```python
# 使用索引选择王微和刘旭阳的英语和艺术课程成绩
selected_scores = scores[(names == '王微') | (names == '刘旭阳'), :][:, (subjects == 'English') | (subjects == 'Art')]

print("王微和刘旭阳的英语和艺术课程成绩:")
print(selected_scores)
```

#### 14、统计scores数组中每名同学所有科目的平均分并显示。

```python
# 沿着列（axis=1）对scores数组进行求和，得到每名同学的平均分
average_scores_per_student = np.mean(scores, axis=1)

print("每名同学所有科目的平均分:")
print(average_scores_per_student)
```
